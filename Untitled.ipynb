### ***Clean 'Last_Name' column by removing any special caracters using strip() method and regular expressions***

```python
data["Last_Name"] = data["Last_Name"].str.strip("123._/")
data
```

### ***Clean Phone_Number columns by adjusting numbers in an unified format after converting data to string to allow slicing, and after removing no matching values of a phone number***

```python
data["Phone_Number"] = data["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)

data["Phone_Number"]=data["Phone_Number"].apply(lambda x: str(x))

data["Phone_Number"] = data["Phone_Number"].apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:])
data["Phone_Number"] = data["Phone_Number"].str.replace("nan--","")
data["Phone_Number"] = data["Phone_Number"].str.replace("Na--","")
data["Phone_Number"]
```

```python
data[["Street_Address","State","Zip_Code"]] = data["Address"].str.split(',',expand=True)
data
```

```python
def clean(x):
    if x == 'N':
        return "No"
    elif x == 'Y':
        return 'Yes'
    else :
        return x

data['Paying Customer'] = data['Paying Customer'].apply(lambda x:clean(x))

```

```python
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('No','N')
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('Yes','Y')
```

```python
data
```

```python
data = data.replace('N/a','')
data = data.fillna('')
data
```

```python
for x in data.index:
    if data.loc[x,"Do_Not_Contact"] == "Y":
        data.drop(x,inplace=True)

data
```

```python
for x in data.index:
    if data.loc[x,"Phone_Number"] == "":
        data.drop(x,inplace=True)

data

# Another way to drop null values 
#data = data.dropin(subset="Phone_Number", inplace=True)
```
